```python
%%writefile app.py

import streamlit as st
import pandas as pd

for_prediction = pd.read_csv("all_predictions_final.csv")

def get_player_info(player_name, year, model):
    player_name = player_name.lower()
    model = model.lower()

    filtered_df = for_prediction[(for_prediction['Player'].str.lower() == player_name) &
                                 (for_prediction['Year'] == int(year)) &
                                 (for_prediction['Model'].str.lower() == model)]

    if not filtered_df.empty:
        result = filtered_df.iloc[0]
        actual_share = result['Share']
        predicted_share = result['predictions']
        actual_rank = result['Rk']
        predicted_rank = result['Predicted_Rk']
    else:
        actual_share = predicted_share = actual_rank = predicted_rank = "No data found"

    return actual_share, predicted_share, actual_rank, predicted_rank

# Streamlit interface
st.title("NBA Player MVP Prediction")
st.write("Enter player's name, year, and model to get the MVP prediction.")

player_name = st.text_input("Enter player's name (e.g., Stephen Curry/stephen curry)")
year = st.number_input("Enter year (2005-2023)", 2005, 2023, step=1)
model = st.selectbox("Select model", ["Ridge", "Random Forest", "Linear", "XGBoost", "SVR", "KNN"])

if st.button("Get Prediction"):
    actual_share, predicted_share, actual_rank, predicted_rank = get_player_info(player_name, year, model)

    st.write("Actual MVP Share:", actual_share)
    st.write("Predicted MVP Share:", predicted_share)
    st.write("Actual MVP Ranking:", actual_rank)
    st.write("Predicted MVP Ranking:", predicted_rank)
```
