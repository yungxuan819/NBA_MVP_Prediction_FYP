```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
mvps.head()
```

```python
players = pd.read_csv("players.csv")
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
players.head()
```

```python
players["Player"].head(50)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex = False)
players.head(50)
```

```python
players.groupby(["Player", "Year"]).get_group(("Alaa Abdelnaby", 1995))
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
players.head(20)
```

```python
players[players["Player"] == "Alaa Abdelnaby"]
```

```python
combined = players.merge(mvps, how = "outer", on=["Player", "Year"])
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
combined
```

```python
teams = pd.read_csv("teams.csv")
teams.head(30)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
teams.head(30)
```

```python
teams.loc[:, "Team"] = teams["Team"].str.replace("*", "", regex=False)
teams.head(5)
```

```python
dataframe1 = pd.read_csv("nicknames.txt")
dataframe1.to_csv("nicknames.csv", index = None)
```

```python
nicknames = pd.read_csv("nicknames.csv")
nicknames
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
combined.head(5)
```

```python
stats = combined.merge(teams, how = 'outer', on = ["Team", "Year"])
stats
```

```python
del stats["Unnamed: 0"]
stats.head()
```

```python
stats.dtypes
```

```python
stats = stats.apply (pd.to_numeric, errors="ignore")
stats.dtypes
```

```python
stats['GB'].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats["GB"].unique()
```
