MSE is useless here as most of the players do not get any mvp votes, so their "Share" will be 0.
Therefore, we will get the top 5 mvp vote getters, and rank them based on accuracy of the ranking.
We will need an error metric (AP) that ranks the players.

```python
combination = combination.sort_values("Share",ascending=False)
combination["Rk"] = list(range(1, combination.shape[0]+1))
combination = combination.sort_values("predictions",ascending=False)
combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
```

```python
combination.head(10)
```

```python
# average precision function
def find_ap(combination):
    actual = combination.sort_values("Share",ascending=False).head(5)
    predicted = combination.sort_values("predictions",ascending=False)
    ps = []
    found = 0
    seen=1
    for index,row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            #print(found)
            #print(seen)
            ps.append(found/seen)
        seen += 1
    return sum(ps)/len(ps)
```

```python
find_ap(combination)
```

To predict every other year (creating backtest function)

```python
def add_ranks(combination):
    combination = combination.sort_values("Share",ascending=False)
    combination["Rk"] = list(range(1, combination.shape[0]+1))
    combination = combination.sort_values("predictions",ascending=False)
    combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
    combination["Diff"] = combination["Rk"] -combination["Predicted_Rk"]
    return combination
```

```python
years = np.arange(1995, 2024)
```

```python
def calculate_dcg(scores):
    dcg = scores[0]
    for i in range(1, len(scores)):
        dcg += scores[i] / np.log2(i + 1)
    return dcg

def find_dcg(combination):
    relevance_scores = combination["Share"].values
    dcg = calculate_dcg(relevance_scores)
    return dcg
```

```python
def backtest(stats, model, years, predictors):
    aps = []
    dcgs = []
    all_predictions = []

    for year in years[5:]:
        train = stats[~(stats["Year"] == year)]
        test = stats[stats["Year"] == year]

        model.fit(train[predictors], train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Year", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))

        dcg = find_dcg(combination)
        dcgs.append(dcg)

    mean_aps = sum(aps) / len(aps)
    mean_dcg = sum(dcgs) / len(dcgs)

    max_dcg = calculate_dcg(sorted(combination["Share"].values, reverse=True))
    normalized_mean_dcg = mean_dcg / max_dcg

    return mean_aps, normalized_mean_dcg, aps, pd.concat(all_predictions)
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=.1)
mean_aps_reg, normalized_mean_dcg_reg, aps, all_predictions_reg = backtest(stats, reg, years[5:],predictors)
```

```python
print("Mean Average Precision:", mean_aps_reg)
print("Normalized Discounted Cumulative Gain:", normalized_mean_dcg_reg)
```

```python
all_predictions_reg
```

```python
def get_share(player_name, year, all_predictions):
    player_name_lower = player_name.lower()
    all_predictions["Player"] = all_predictions["Player"].str.lower()
    player_data = all_predictions.loc[(all_predictions["Player"] == player_name_lower) & (stats["Year"] == year)]
    if player_data.empty:
        return "Invalid"
    else:
        actual_share = player_data["Share"].values[0]
        predicted_share = player_data["predictions"].values[0]
        actual_rank = player_data["Rk"].values[0] if "Rk" in player_data.columns else "N/A"
        predicted_rank = player_data["Predicted_Rk"].values[0] if "Predicted_Rk" in player_data.columns else "N/A"
        return actual_share, predicted_share, actual_rank, predicted_rank
```

```python
player_name = input("Enter player's name: ")
year = int(input("Enter the year: "))

actual_share, predicted_share, actual_rank, predicted_rank = get_share(player_name, year, all_predictions_reg)

if actual_share == 0:
    print("Invalid input. The player either doesn't exist or didn't receive MVP votes that year.")
else:
    print("Actual MVP Share:", actual_share)
    print("Predicted MVP Share:", predicted_share)
    print("Actual Ranking:", actual_rank)
    print("Predicted Ranking:", predicted_rank)
```
