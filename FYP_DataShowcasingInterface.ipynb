```python
%%writefile app.py

import pandas as pd
import streamlit as st

stats = pd.read_csv("players_mvp_stats.csv")

def get_player_stats(player_name, year, statistics):
  """
  This function retrieves player statistics based on input parameters.

  Inputs:
      player_name (str): Name of the player (uppercase/lowercase)
      year (int): A specific year
      statistics (str): Name of the desired statistic (uppercase/lowercase)

  Output:
      str: Formatted string containing the requested stat or error message
  """

  # Convert player_name and column_name to lowercase for case-insensitive matching
  player_name = player_name.lower()
  statistics = statistics.lower()

  # Filter DataFrame based on input
  player_stats = stats[(stats['Player'].str.lower() == player_name) & (stats['Year'] == year)]

  # Check if the player and year combination exists
  if player_stats.empty:
    return f"No data found for {player_name} in {year}"

  # Check if the requested stats column exists
  matching_columns = [col for col in player_stats.columns if col.lower() == statistics]
  if not matching_columns:
    return f"Invalid column name: {statistics}"

  # Get the specific stats for the player
  stat_value = player_stats[matching_columns[0]].values[0]

  return f"{player_name}'s average {statistics} in {year}: {stat_value}"

# Streamlit App

st.title("Player Statistics Explorer")

player_name = st.text_input("Enter Player Name")
year = st.number_input("Select Year", min_value=min(stats["Year"]), max_value=max(stats["Year"]))
statistics = st.text_input("Enter Statistic Name (enter abbreviations)")

if st.button("Get Stats"):
  output_text = get_player_stats(player_name, year, statistics)
  st.write(output_text)
```
